```python
# Using the Spacy praser to get the basic dependency relationship and head
doc = nlp(text)

for token in nlp(text):
    print(token.text,'=>',token.dep_,'=>',token.head.text)
    
```

```python
#Table form: Give the POS tag info
Parser_doc = nlp(text)
for token in Parser_doc:
    print(
        f""" 
    TOKEN: {str(token)}
    =====
    TAG: {str(token.tag_):10} POS: {token.pos_}
    EXPLANATION: {spacy.explain(token.tag_)}""")
```

```python


doc = nlp(text)
    
def Indirect_Quote_PVP(nlp_doc):
    pattern = [{"POS": "PROPN"},{"POS": "VERB"}, {"POS": "PRON"}, {"ORTH":{"IN": [".","!", "?"]}}] #{"ORTH":{"IN": [".","!", "?"]}}]
    matcher.add("IndirectQ_PVP", [pattern], greedy = 'LONGEST')
   
    matches = matcher(nlp_doc)

    print(len(matches))
    return matches

matches_o = Indirect_Quote_PVP(doc)
    

for Indirect_Quote_PVP, start, end in matches_o:
    print(doc[start:end])
    
 #Okay, I think I manage to get the first half of sentence...but...  
            
```
